```python
# Analyzing Chicago Red Light Camera Data
import pandas as pd
%matplotlib inline
```

```python
# Questions
# 1. How much money does the city make from red light camera tickets every year
# 2. Which red light camera stops are the most revenue generators?
# 3. Does weather influence the amount of violations?
# 4. What day is the most likely to get a red light camera violation
# 5. What is the overall trend of red light camera violations? Up or down over time
```

```python
# Read in the data frame
Red_Light_Data = pd.read_csv('Red_Light_Camera_Violations.tsv',sep='\t')
Red_Light_Data.head(5)
```

```python
# Reassign the Date to a Date Time
Red_Light_Data['Date_Time'] = pd.to_datetime(Red_Light_Data['VIOLATION DATE'])
```

```python
# Read in Some Chunks of Data
RL_Data_2015 = Red_Light_Data[(Red_Light_Data['Date_Time'] >= '01/01/2015') &
                       (Red_Light_Data['Date_Time'] <= '12/31/2015')]
RL_Data_2016 = Red_Light_Data[(Red_Light_Data['Date_Time'] >= '01/01/2016') &
                       (Red_Light_Data['Date_Time'] <= '12/31/2016')]
RL_Data_2017 = Red_Light_Data[(Red_Light_Data['Date_Time'] >= '01/01/2017') &
                       (Red_Light_Data['Date_Time'] <= '12/31/2017')]
```

```python
# Focus on 2017
RL_Data_2017_Group = RL_Data_2017.groupby(['Date_Time'])
pd.DataFrame(RL_Data_2017_Group['VIOLATIONS'].sum()).plot()
```

```python
# Get statistics on each camera location
# I.e. Which cameras are the most revenue generating

RL_Data_2017_Camera_Group = RL_Data_2017.groupby(['LOCATION'])
Cameras = pd.DataFrame(RL_Data_2017_Camera_Group['VIOLATIONS'].sum())
Cameras.sort_values(ascending=False,by='VIOLATIONS',inplace=True)
Cameras.head()
Cameras.reset_index(level=0, inplace=True)
```

```python
import math
from ast import literal_eval
def merc(Coords):
    Coordinates = literal_eval(Coords)
    lat = Coordinates[0]
    lon = Coordinates[1]
    
    r_major = 6378137.000
    x = r_major * math.radians(lon)
    scale = x/lon
    y = 180.0/math.pi * math.log(math.tan(math.pi/4.0 + 
        lat * (math.pi/180.0)/2.0)) * scale
    return (x, y)
```

```python
Cameras['coords_x'] = Cameras['LOCATION'].apply(lambda x: merc(x)[0])
Cameras['coords_y'] = Cameras['LOCATION'].apply(lambda x: merc(x)[1])
Cameras['circle_sizes'] = Cameras['VIOLATIONS'] / 300
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.tile_providers import CARTODBPOSITRON_RETINA
from bokeh.models import ColumnDataSource, HoverTool, WheelZoomTool
from bokeh.io import save
source = ColumnDataSource(data=dict(
                        x=list(Cameras['coords_x']), 
                        y=list(Cameras['coords_y']),
                        cost=list(Cameras['VIOLATIONS']*100),
                        sizes=list(Cameras['circle_sizes'])))
hover = HoverTool(tooltips=[
    ("Yearly Revenue","$@cost{0,0}")
    
])
# range bounds supplied in web mercator coordinates
p = figure(x_axis_type="mercator", 
           y_axis_type="mercator",plot_width=650, plot_height=800,tools=['pan'],title="Revenue by Red Light Camera - Chicago - 2017")
p.add_tile(CARTODBPOSITRON_RETINA)
p.circle(x='x',
         y='y', 
         source=source,
         size='sizes',
         line_color="#FF0000", 
         fill_color="#FF0000",
         fill_alpha=0.2,hover_alpha=0.6,hover_fill_color="green")

p.toolbar.active_scroll = "auto"
wheel = WheelZoomTool()
p.add_tools(wheel)
p.add_tools(hover)
p.toolbar.active_scroll = wheel 

p.toolbar.logo = None

p.xaxis.major_tick_line_color = None  # turn off x-axis major ticks
p.xaxis.minor_tick_line_color = None  # turn off x-axis minor ticks

p.yaxis.major_tick_line_color = None  # turn off y-axis major ticks
p.yaxis.minor_tick_line_color = None  # turn off y-axis minor ticks

p.xaxis.major_label_text_font_size = '0pt'  # turn off x-axis tick labels
p.yaxis.major_label_text_font_size = '0pt'  # turn off y-axis tick labels
    
output_notebook()
save(p)
```
